```python
#Black Jack Game with Pyhton using OOPs and error handling Concept

import random


class DeckSelection():
    
    #suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
    
    crdnumber =('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
    
    deck = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,'Queen':10, 'King':10, 'Ace':11}
    
    #decknumber = random.choice(crdnumber)
    
    playing = True
```

```python
class Players(DeckSelection):
    
    plcount=0
    dlcount=0
    playerrandomcard,dealerrandomcard,playeranotherrandomcard,dealeranotherrandomcard=0,0,0,0
    playercard1,playercard2,dealercard1,dealercard2,sumplayercards,sumdealercards=0,0,0,0,0,0
     
           
    def playercards(self):
        
        try:
       
            player = input('Do want to play black jack(y/n): ').lower()

            if 'y' in player:

                d=DeckSelection()

                Players.playerrandomcard= random.choice(d.crdnumber)

                Players.dealerrandomcard = random.choice(d.crdnumber)

                Players.plcount+=1
                Players.dlcount+=1


                print('first Card given to player: {}' .format(Players.playerrandomcard))
                print('first Card given to dealer: {}' .format(Players.dealerrandomcard))

                while Players.plcount <= 1 and Players.dlcount <= 1:

                    #pick another card

                    Players.playeranotherrandomcard= random.choice(d.crdnumber)

                    Players.plcount+=1

                    print('Second Card given to player: {}' .format(Players.playeranotherrandomcard))

                    Players.dealeranotherrandomcard = random.choice(d.crdnumber)

                    Players.dlcount+=1

                    print('Second Card given to dealer: {}' .format(Players.dealeranotherrandomcard))

                    Players.playercard1=int(d.deck[Players.playerrandomcard])
                    print("Numeric value of players card1: {}" .format(Players.playercard1))
                    Players.playercard2=int(d.deck[Players.playeranotherrandomcard])
                    print("Numeric value of players card2: {}" .format(Players.playercard2))

                    Players.sumplayercards = Players.playercard1 + Players.playercard2

                    print("Sum of cards present at the player: {}" .format(Players.sumplayercards))    

                    Players.dealercard1=int(d.deck[Players.dealerrandomcard])
                    print("Numeric value of dealer card1: {}" .format(Players.dealercard1))
                    Players.dealercard2=int(d.deck[Players.dealeranotherrandomcard])
                    print("Numeric value of dealer card2: {}" .format(Players.dealercard2))

                    Players.sumdealercards = Players.dealercard1 + Players.dealercard2
                    print("Sum of cards present at the dealer: {}" .format(Players.sumdealercards))

            return Players.sumplayercards,Players.sumdealercards,Players.playerrandomcard,Players.playeranotherrandomcard,Players.dealerrandomcard,Players.dealeranotherrandomcard
        
        
        except:
            
            print('Error occured')
            #continue
            
        else:
            
            print('Check exception')
            #break
    
                
    
```
